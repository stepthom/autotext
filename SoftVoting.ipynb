```python
import pandas as pd
df = pd.read_csv('../data/pump_test.csv')
df.head()
```

```python
others = [
        # LGBM, 50 mestimate, 0.8257
        '92b4dd21-0931-47e9-a983-da87d253d70a-probas.csv',

        # Stacked (lgbm, rf, hist), 100 mestimate,0.8257
        'e46cb09e-bdd2-4e5a-aace-2c661ede3d75-probas.csv',
    
        # RF, 50 mestimate, 0.8255
        '81d1551b-d488-4ba9-a617-f203466ee726-probas.csv',
    
        #FLAML LGBM+RF+XGB ensemble=True, 75 mestiamte, 0.8211
        'cb63c4cf-f41b-40b0-9f88-4589d324a1e5-probas.csv',
         ]

colclasses = {}
classes = ['functional', 'functional needs repair', 'non functional']
for c in classes:
    colclasses[c] = []
p = 0
for other in others:
    _df = pd.read_csv('../out/' + other)
    for c in classes:
        colname = 'p' + str(p) + "_" + c
        df[colname] = _df[c]
        colclasses[c].append(colname)
    p = p + 1
    
pcols = ["p{}".format(i) for i in range(p)]
```

```python
# sum them up!
for cl in colclasses:
    print(cl)
    print(colclasses[cl])
    df[cl] = df[colclasses[cl]].sum(axis=1)
    
df['status_group'] = df[classes].idxmax(axis=1)
```
