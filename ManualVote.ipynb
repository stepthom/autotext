```python
def matches(row):
    count = 0
    for c in pcols:
        if c != 'p0' and row['p0'] == row[c]:
            count = count+1
    return count

df['p0_matches'] = df.apply(matches, axis=1)
```

```python
def get_label(row):
    if row['p0'] == row['p1']:
        return row['p0']
    if row['p0_matches'] >= 2:
        return row['p0']
    else:
        return row['p1']
```

```python
def mymode(row):
    return row.mode().iloc[0]

df['mode'] = df[pcols].apply(mymode, axis=1)
```

```python
df['status_group'] = df['mode']
df[['id', 'status_group']].to_csv("out/manual_6mode-stepthom_submission.csv", index=False)
```
