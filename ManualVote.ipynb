```python
others = ['d5881dc8-2ca7-4ca6-82db-b67960ae93c6-stepthom_submission.csv',
          '0baf47ac-3395-4a2b-8dc1-6e56fdc05209-stepthom_submission.csv',
          'f06f31fa-e145-42f5-8a77-dd7a0818e747-stepthom_submission.csv',
          #'9ac1099b-baa8-406d-8e1a-9b4fd2bfa470-stepthom_submission.csv',
          
          
          
          'bab2166e-66ae-4e2a-85a2-fbb7d916e680-stepthom_submission.csv',
          'f18637f2-699e-47cd-b55d-0e8acdd1d2c8-stepthom_submission.csv',
          'ak_100_None_30_15-stepthom_submission.csv'
          #'-stepthom_submission.csv',
          #'-stepthom_submission.csv',
          
         ]

p = 0
for other in others:
    _df = pd.read_csv('out/' + other)
    df['p' + str(p)] = _df.status_group
    p = p + 1
    
pcols = ["p{}".format(i) for i in range(p)]
```

```python
def matches(row):
    count = 0
    for c in pcols:
        if c != 'p0' and row['p0'] == row[c]:
            count = count+1
    return count

df['p0_matches'] = df.apply(matches, axis=1)
```

```python
def get_label(row):
    if row['p0'] == row['p1']:
        return row['p0']
    if row['p0_matches'] >= 2:
        return row['p0']
    else:
        return row['p1']
```

```python
df['status_group'] = df.apply(get_label, axis=1)
df.head(40)
```
